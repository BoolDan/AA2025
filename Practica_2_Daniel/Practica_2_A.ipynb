```python
# --- PARTE A: Regresi√≥n Lineal ---
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

def generarDatosFC():
    np.random.seed(0)
    x = np.arange(1, 61).reshape(-1, 1)
    y = 0.7 * x.flatten() + 60 + np.random.uniform(-5.9, 5.9, size=x.size)
    return x, y

x, y = generarDatosFC()

# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(x, y)

# Predicciones
y_pred = modelo.predict(x)

# Mostrar resultados
print("Coeficientes:", modelo.coef_)
print("Intercepto:", modelo.intercept_)

# Gr√°fica
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, y_pred, color='red', linewidth=2, label='Regresi√≥n lineal')
plt.xlabel('Tiempo (s)')
plt.ylabel('Frecuencia Cardiaca (FC)')
plt.legend()
plt.show()
```
